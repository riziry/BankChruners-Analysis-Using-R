```
pie(table(da$Education_Level[da$Attrition_Flag=="Attrited Customer"]), main="Pie of educational_level on Attrited Customer")
cat("\nEducational_level-Attrited\n")
print(count(table(da$Education_Level[da$Attrition_Flag=="Attrited Customer"])))
```

First we want to know the **characteristics of educational level** from the attrited customer.

Then from there we can see that **the most** attrited customers are customers who have educational level **graduate** with a total of 487 people.

```
pie(table(da$Card_Category[da$Attrition_Flag=="Attrited Customer" & da$Education_Level=="Graduate"]), main="Pie of Card_Category Customer that Attrited and Graduate")
cat("\nCard Category-Attrited-Graduate\n")
print(count(table(da$Card_Category[da$Attrition_Flag=="Attrited Customer" & da$Education_Level=="Graduate"])))
```

And from this we get the most **attried and graduate** is having a **blue** card with a total 450 people

```
pie(table(da$Income_Category[da$Attrition_Flag=="Attrited Customer" & da$Education_Level=="Graduate" & da$Card_Category=="Blue"]), main="Pie of Income_Category Customer that Attrited, Graduate\nwith Blue Card")
cat("\nIncome Category-Attrited-Graduate-Blue Card\n")
print(count(table(da$Income_Category[da$Attrition_Flag=="Attrited Customer" & da$Education_Level=="Graduate" & da$Card_Category=="Blue"])))
```

And from the graph we can see the **less than $40K** income category is the most income category in customers who are **attrited, graduate and have a blue card**

```
pie(table(da$Marital_Status[da$Attrition_Flag=="Attrited Customer" & da$Education_Level=="Graduate" & da$Card_Category=="Blue" & da$Income_Category=="Less than $40K"]), main="Pie of Marital_Status Customer that Attrited, Graduate\nwith Blue Card and have income less than $40k")
cat("\nMarital_Status-Attrited-Graduate-Blue Card-Less than $40k\n")
print(count(table(da$Marital_Status[da$Attrition_Flag=="Attrited Customer" & da$Education_Level=="Graduate" & da$Card_Category=="Blue" & da$Income_Category=="Less than $40K"])))
```

and now at the **marital status** we cant just see at the graph to know it because **the single and the married** is have amount kinda **similar**, but we can see it from the count function that we use and get **the single** are **have more** customers than the married, with total 85 people

```
pie(table(da$Gender[da$Attrition_Flag=="Attrited Customer" & da$Education_Level=="Graduate" & da$Card_Category=="Blue" & da$Income_Category=="Less than $40K" & da$Marital_Status=="Single"]), main="Pie of Gender Customer that Attrited, Graduate\nwith Blue Card, have income less than $40k and Single")
cat("\nGender-Attrited-Graduate-Blue Card-\nLess than $40k-Marital_Status\n\n")
print(count(table(da$Gender[da$Attrition_Flag=="Attrited Customer" & da$Education_Level=="Graduate" & da$Card_Category=="Blue" & da$Income_Category=="Less than $40K" & da$Marital_Status=="Single"])))
```

from that we can see clearly that the **"F" or Female** are have **more amount of customer** that **attrited**, **graduate**, use **blue** card, have income ****less than $40K**, and **single**
